```python
with open('sowpods.txt', 'r') as Words:
    string = Words.read()
    sowpods = (string.split('\n'))
    sowpods.remove('')
    sowpods = {str(i) for i in sowpods if len(i) <= 9}
#converts the SOWPODS scrabble dictionary into a list
```

```python
import random


vowels = ['a', 'e', 'i', 'o', 'u']
consonants = ['b', 'c', 'd', 'f', 'g', 'h', 'j', 
               'k', 'l', 'm', 'n', 'p', 'q', 'r', 
               's', 't', 'v', 'w', 'x', 'y', 'z']
```

```python
def create_board(vowel_number):
    '''creates a countdown board of 9 letters and returns the largest word present'''
    board = []

    for v in range (vowel_number):
        board.append(vowels[random.randint(0, 4)])
    for c in range (9 - vowel_number):
        board.append(constonants[random.randint(0, 20)])
    return board
```

```python
def solve_board(vowel_number, display_board = False):
    '''creates a countdown board of 9 letters and returns the largest word present'''
    
    board = ['e', 'o', 'a', 'm', 'c', 't', 'c', 'w', 'y']
    length = []
        
    #creates list of all words that can be created from the board letters,
    #finds their lengths and returns stats on the word 
    for word in sowpods:
        testboard = list(board)
        count = 0
        for letter in word:
            if letter in testboard:
                testboard.remove(letter)
                count += 1
        if len(word) == count:
            length.append(len(word))
    if len(length) == 0:
        return 0
    return max(length)
```

```python
import itertools 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def powerset(iterable):
    """
    This is a modification of the recipe available here: 
    https://docs.python.org/3/library/itertools.html#itertools-recipes
    
    Note that the main modification is to ensure the order is of the longest subset first:
    
    powerset([1,2,3]) --> (1,2,3) (1, 2) (1, 3) (2, 3) (1,) (2,) (3,)"
    """
    s = list(iterable)
    return itertools.chain.from_iterable(itertools.combinations(s, r) for r in range(len(s)+1, 1, -1))

vowels = 'aeiou'
consonants = 'bcdfghjklmnpqrstvwxyz'

dictionary = set([tuple(sorted(word)) for word in sowpods])

def solve_board_vince(
    number_of_vowels, 
    display_board=False, 
    known_words_as_a_set=dictionary, 
    vowels=vowels, 
    consonants=consonants,
    board_size=9,
    seed=None,
):
    """creates a countdown board of 9 letters and returns the largest word present"""

    if seed is not None:
        np.random.seed(seed)
        
    #randomly generates the board
    board = list(np.random.choice(list(vowels), number_of_vowels, replace=True))
    board += list(np.random.choice(list(consonants), board_size - number_of_vowels, replace=True))


    if display_board is True:
        print(board)
    
    sorted_board = tuple(sorted(board))
    
    for subset in powerset(sorted_board):
        if subset in known_words_as_a_set:
            return subset
    return tuple([])
```
