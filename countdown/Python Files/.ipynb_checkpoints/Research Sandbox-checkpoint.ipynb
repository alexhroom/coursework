```python
with open('sowpods.txt', 'r') as Words:
    string = Words.read()
    sowpods = (string.split('\n'))
    sowpods.remove('')
    sowpods = [str(i) for i in sowpods if len(i) <= 9]
#converts the SOWPODS scrabble dictionary into a list
```

```python
import random


vowels = ['a', 'e', 'i', 'o', 'u']
constonants = ['b', 'c', 'd', 'f', 'g', 'h', 'j', 
               'k', 'l', 'm', 'n', 'p', 'q', 'r', 
               's', 't', 'v', 'w', 'x', 'y', 'z']
```

```python
def create_board(vowel_number):
    '''creates a countdown board of 9 letters and returns the largest word present'''
    board = []

    for v in range (vowel_number):
        board.append(vowels[random.randint(0, 4)])
    for c in range (9 - vowel_number):
        board.append(constonants[random.randint(0, 20)])
    return board
```

```python
def solve_board(vowel_number, display_board = False):
    '''creates a countdown board of 9 letters and returns the largest word present'''
    
    board = ['e', 'o', 'a', 'm', 'c', 't', 'c', 'w', 'y']
    length = []
        
    #creates list of all words that can be created from the board letters,
    #finds their lengths and returns stats on the word 
    for word in sowpods:
        testboard = list(board)
        count = 0
        for letter in word:
            if letter in testboard:
                testboard.remove(letter)
                count += 1
        if len(word) == count:
            length.append(len(word))
    if len(length) == 0:
        return 0
    return max(length)
```

```python
count = 0
board = create_board(3)
print(board)
for letter in list("aaaaaaaaaaaaaaaaaaaaa"):
    if letter in board:
        board.remove(letter)
        count += 1
print(board)
print(count)
```

```python
from collections import Counter

def solve_board_coll(vowel_number, display_board = False):
    '''creates a countdown board of 9 letters and returns the largest word present'''

        
    board = ['e', 'o', 'a', 'm', 'c', 't', 'c', 'w', 'y']
    length = []
        
    #creates list of all words that can be created from the board letters,
    #finds their lengths and returns stats on the word 

    testboard = Counter(board)
    length = []
    for word in sowpods:
        if not Counter(word) - testboard:
            break
        else:
            length.append(len(word))
            print(word)
    return max(length)
```

```python
def solve_board_negative(vowel_number, display_board = False):
    '''creates a countdown board of 9 letters and returns the largest word present'''

        
    board = ['e', 'o', 'a', 'm', 'c', 't', 'c', 'w', 'y']
    length = []
        
    for word in sowpods:    
        on_board = True
        testboard = list(board)
        for letter in word:
            if letter not in list(testboard):
                on_board = False
                break
            else:
                testboard.remove(letter)
        if on_board == True:
            length.append(len(word))
    return length
```
